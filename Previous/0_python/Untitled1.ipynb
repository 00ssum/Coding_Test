```python
graph = dict()

graph['A'] = ['B', 'C']
graph['B'] = ['A', 'D']
graph['C'] = ['A', 'G', 'H', 'I']
graph['D'] = ['B', 'E', 'F']
graph['E'] = ['D']
graph['F'] = ['D']
graph['G'] = ['C']
graph['H'] = ['C']
graph['I'] = ['C', 'J']
graph['J'] = ['I']

print(graph)
```

```python
def bfs(graph, start_node):
    visited = []
    queue = [start_node]
    answer = 0

    while queue:
        node = queue.pop(0)
        if node in visited:
            continue
        for i in graph[node]:
            if i not in visited:
                queue.append(i)
                answer += 1
        visited.append(node)
    print(answer)
    print(visited)

bfs(graph, 'A')
```

```python
def dfs(graph, start_node):
    visited = []
    queue = [start_node]

    while queue:
        node = queue.pop()
        if node not in visited:
            queue.extend(sorted(graph[node], reverse=True))
            visited.append(node)
    print(visited)

dfs(graph, 'A')
```

```python
import heapq

def dijkstra(graph, start_node):
    distanses = {node: float('inf') for node in graph.keys()}
    distanses[start_node] = 0
    queue = []
    heapq.heappush(queue, [distanses[start_node], start_node])

    while queue:
        dist, node = heapq.heappop(queue)
        if distanses[node] < dist:
            continue
        
        for item in graph[node].items():
            nnode, ndist = item
            if ndist + dist < distanses[nnode]:
                distanses[nnode] = ndist + dist
                heapq.heappush(queue, [distanses[nnode], nnode])
    print(distanses)

dijkstra(graph_d, 'A')
```

```python
# 크루스칼
def find(parent, a):
    if parent[a] != a:
        parent[a] = find(parent, parent[a])
    return parent[a]

def union(parent, a, b):
    a = find(parent, a)
    b = find(parent, b)

    if a > b:
        parent[a] = b
    else:
        parent[b] = a

def solution(edges, n):
    edges.sort(key=lambda x: x[2])
    print(edges)
    parent = [i for i in range(n+1)]
    answer = 0

    for a, b, cost in edges:
         if find(parent, a) != find(parent, b):
             answer += cost
             union(parent, a, b)
    print(answer)


edges = [(1,2,13), (1,3,5), (2,4,9), (3,4,15), (3,5,3), (4,5,1), (4,6,7), (5,6,2)]
solution(edges, 6)
```
