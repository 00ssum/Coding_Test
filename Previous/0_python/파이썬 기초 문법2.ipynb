#1. Python Data Structure
## 1.1 스택(Stack)
 - FILO
 - List를 활용해서 구현이 가능

```python
a = [1, 2, 3, 4, 5]
a.append(10)
a.append(20)
print(a)
c = a.pop() # 마지막 값이 꺼내어 진다. FILO
print(c)
print(a)
```

## 1.2 큐 (Queue)
- FIFO
- List를 활용해서 구현이 가능


```python
a = [1, 2, 3, 4, 5]
a.append(10)

c = a.pop(0)
print(c)
print(a)
```

## 1.4 집합 (Set)
> - 중복 없는 값의 데이터 집합
- 순서가 없다

```python
s = {1, 2, 3, 1, 2, 3}
print(s)
s = set([1, 2, 3, 1, 2, 3])
print(s)

s.add(4)
print(s)
s.remove(4)
s.discard(1)
print(s)
s.update([2,3,4,5,6])
print(s)

print('----------------------')
s1 = {1,2,3,4,5}
s2 = {3,4,5,6,7}

print(s1.union(s2))        # 합집합
print(s1.intersection(s2)) # 교집합
print(s1 & s2)             # 교집합
print(s1.difference(s2))   # 차집합
```

## 1.5 사전 (Dictionary) == Map<Key, Value>
>- Key, Value로 구분할 수 있는 데이터 구조

```python
student = {}
student = {'Kim': 23, 'Lee': 30, 'Park': 18, 'Choi': 50}
print(student)
print(student.items())

for dict_items in student.items():
    print(type(dict_items))

# 키 / 값 출력
print(student.keys())
print(student.values())

for k,v in student.items():
    print(f'key : {k}')
    print(f'values : {v}')

# Add Data
student['Jo'] = 10
print(student)
# Remove Data
student.pop('Jo')
print(student)
```

## 1.6 Collection 모듈
### 1.6.1 Deque
>- Linked List
- Rotate, reverse

```python
from collections import deque

deque_list = deque()
for i in range(5):
    deque_list.append(i)
print(deque_list)
deque_list.appendleft(10)
print(deque_list)
deque_list.rotate(2)
print(deque_list)
deque_list.rotate(-2)
print(deque_list)

deque_list.extend([5,6,7])
print(deque_list)

deque_list.pop()
print(deque_list)
```

## 2.2 list comprehension
- 기존 List를 사용하여 다른 List를 만드는 기법
- 포괄적인 List, 포함되는 리스트라는 의미

```python
# for 문
result = [i for i in range(10)]
result

# 이중 for 문 - 1
list1 = ['a', 'b', 'c']
list2 = ['1', '2', '3']

result = [i+j for i in list1 for j in list2]
print(result)

# filter 조건문
result = [i+j for i in list1 for j in list2 if not(j=='3')]
print(result)

result = [i+j if i=='a' else i for i in list1 for j in list2 if not(j=='3')]
print(result)


items = "zero one two three"
list_items = items.split()
print(list_items)
print([[w.upper(), w.lower(), w] for w in list_items])

# 이중 for문 - 2
list1 = ['a', 'b', 'c']
list2 = ['1', '2', '3']

result = [[i+j for i in list1] for j in list2]
print(result)
```

## 2.4 lambda & map & reduce

### 2.4.1 lambda
- 익명함수

```python
f = (lambda x, y : x+y)
f(5,7)

up_low = lambda x: x.upper() + x.lower()
up_low('in')
```

2.4.2 Map
- 두 개 이상의 list에도 적용 가능

```python
ex = [1,2,3,4,5]
f = lambda x: x**2

print(list(map(f, ex)))
print([f(v) for v in ex])

print([f(v) if v%2 == 0 else v for v in ex ])
```
