```python
import re


st = '...!@BaT#*..y.ab-_cdefghijklm'

st = st.lower()
print(st)
st = re.sub('[^a-z0-9\-_.]', '', st)
print(st)
st = re.sub('\.+', '.', st)
print(st)
st = re.sub('^[.]|[.]$', '', st)
print(st)
st = 'a' if len(st) == 0 else st[:15]
print(st)
st = re.sub('^[.]|[.]$', '', st)
print(st)
st = st if len(st) > 2 else st + "".join([st[-1] for i in range(3-len(st))])

print(st)
```

### match(pattern, string, flags)
re.match() 함수는 "문자열의 처음" 부터 시작하여 패턴이 일치되는 것이 있는지를 확인한다.

```python
import re
matchObj = re.match('a', 'a')
print(matchObj)

print(re.match('a', 'abc'))
print(re.match('a', 'cba'))
print(re.match('a', 'ccc'))

# a 로 시작하는 것을 찾는다.
```

매치되지 않을 때 re.match 함수는 None 을 반환한다.

### search(pattern, string, flags)
re.search() 함수는 "문자열 안"에 일치하는 값이 있는지를 확인한다. (처음부터 같을 필요가 없다)

```python
import re
matchObj = re.search('a', 'a')
print(matchObj)

print(re.search('a', 'abc'))
print(re.search('a', 'cba'))
print(re.search('a', 'ccc'))
```

### findall(patter, string, flags)
re.findall() 함수는 일치하는 모든 부분을 리스트로 반환한다.

```python
import re
matchObj = re.findall('a', 'a')
print(matchObj)

print(re.findall('a', 'abc'))
print(re.findall('a', 'cbaa'))
print(re.findall('a', 'caba'))
print(re.findall('a', 'ccc'))
print(re.findall('aaa', 'aaaaa'))
```

### match Object의 메서드
- group() : 일치된 문자열을 반환
- start() : 일치된 문자열의 시작위치를 반환
- end()   : 일치된 문자열의 끝 위치를 반환
- span()  : 일치된 문자열의 (시작, 끝)위치를 반환





```python
import re

matchObj = re.search('match', "'matchObj' is a good name, but 'm' is convenient.")
print(matchObj)
print(matchObj.group()) # None 이 아니어야함
print(matchObj.start())
print(matchObj.end())
print(matchObj.span())
```

### 대괄호 : 여러 문자 중 하나와 일치


```python
matchObj = re.search('Why [a-z]o serious\?', 'Why so serious?')
print(matchObj)
matchObj = re.search('Why [0-9]o serious\?', 'Why so serious?')
print(matchObj)
# ^ not
matchObj = re.search('Why [^0-9]o serious\?', 'Why so serious?')
print(matchObj)

# . 모든 문자
matchObj = re.search('Why .o serious\?', 'Why so serious?')
print(matchObj)
```

### sub(pattern, repl, string, count, flag)


```python
import re

print(re.sub('\d{4}', 'xxxx', '010-1234-5678'))
print(re.sub('\d{4}', 'xxxx', '010-1234-5678', 1))
```

```python
# 문자열 빈칸 채우기
a = 'a'

# rjust() : 오른쪽에 두고 왼쪽 채우기 / ljust() : 왼쪽에 두고 오른쪽 채우기
print(a.rjust(10, ' '))
print(a.ljust(10, ' '))

a = a.rjust(10, ' ')
print(a)
print(a.lstrip())
a = 'a'
a = a.ljust(10, ' ')
print(a)
# 공백 제거하기 lstrip() : 왼쪽 공백 / rstrip() : 오른쪽 공백
print(a.lstrip())

num = str(10)
print(num.rjust(10, '0'))
```
