## 1.1. 객체지향 프로그래밍 개요
>- 객체: 실생활에서 일종의 물건 ->
속성(Attribute)와 행동(Action)을 가짐




```python
class HighSchoolStudent(object):
    def __init__(self, name, id, class_room):
       self.name = name
       self.id = id
       self.class_room = class_room
    
    def change_class_room(class_room):
        self.class_room = class_room

    def __str__(self):
        return '123'

```

```python
s1 = HighSchoolStudent('Kim', '20210605', '5')
print(s1)
```
