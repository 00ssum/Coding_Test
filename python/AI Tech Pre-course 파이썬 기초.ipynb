# 2.3 Conditionals and Loops


## 2.3.1 Conditionals

```python
print('-------------------주소 비교문 is-------------------')
# 주소 비교문 is
a = 100
b = 100
print(a == b)
print(a is b)

# -5 ~ 256 까지는 미리 만들어둠
a = 300
b = 300
print(a is b)

print('-------------------존재하면 참 아니면 거짓--------------------')

# 존재하면 참 아니면 거짓
if '1':
    print('hello')
else:
    print('bye')

if 0:
    print('hello')
else:
    print('bye')

print('--------------------Boolean List -------------------')
# Boolean List 
boolean_list = [True, True, True, True, False, True]
print(all(boolean_list))
print(any(boolean_list))

print('---------------------삼항연산자------------------------')
# 삼항연산자
values = 12
is_even = True if values & 2 == 0 else False
print(is_even)

print('----------------------------------------------------------')
```

## 2.3.2 Loop
